```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
iris=load_iris()

X=pd.DataFrame(iris.data,columns=X.columns)
y=iris.target
```

```python
X.head()
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
X_scaled.shape
```

```python
# PCA
from sklearn.decomposition import PCA
# 2. Instantiate PCA object (e.g., reduce to 2 components for visualization)
pca = PCA(n_components=2)

# 3. Fit and transform the scaled data
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca.shape
```

```python
print("Explained Varence Ratio :",pca.explained_variance_ratio_)
```

```python
print(pca.components_)
```

```python
plt.figure(figsize=(7,7))
plt.scatter(
    X_pca[:,0],
    X_pca[:,1],
    c=y
)
plt.title("PCA for iris data")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
```
